```python

import matplotlib.pyplot as plt

# Data
currencies = ["EUR", "GBP", "JPY", "AUD", "MXN", "CAD", "KRW"]
returns = [-9.3, 4.0, 1.9, -0.9, -9.47, 0.9, 5.47]
ppp = [36.0, 31.0, -41.2, -9.8, 17.7, 0.4, -25.0]

plt.figure(figsize=(8,6))
plt.scatter(returns, ppp)

for x, y, label in zip(returns, ppp, currencies):
    plt.text(x, y, label, fontsize=10)

plt.xlabel("1-year USD return (%)")
plt.ylabel("PPP misalignment (%)")
plt.title("Momentum vs GDP-adjusted Big Mac PPP Across Selected Currencies")

plt.tight_layout()

output_path = "/mnt/data/momentum_vs_ppp_scatter.png"
plt.savefig(output_path, dpi=300)

output_path
```
